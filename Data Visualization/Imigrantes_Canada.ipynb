Importando a Biblioteca Pandas

```python
import pandas as pd
```

Importando dados da planilha


```python
df = pd.read_csv('imigrantes_canada.csv')
df
```

Informações dos dados

```python
df.info()
```

Alocando a coluna País como Index dos dados

```python
df.set_index('País', inplace=True)
```

Gerando uma lista dos anos de 1980 a 2013

```python
anos = list(map(str, range(1980, 2014)))
anos
```

Localizando somente o Pais Brasil

```python
brasil = df.loc['Brasil', anos]
brasil
```

Criando um DataFrame Series

```python
brasil_dict = {'Ano': brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

Praticando: Criando um DataFrame de um dicionario para as vendas dos meses de 4 diferentes lojas

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

dados_vendas = pd.DataFrame(vendas_2022, index=lojas)
dados_vendas
```

Praticando: Criando uma figura com subplots para comparar as vendas das lojas

```python
fig, axs = plt.subplots(2, 2, figsize=(14,8))
fig.subplots_adjust(hspace=0.3, wspace=0.4)
fig.suptitle('Vendas mensais de 2022 das lojas A, B, C e D')

axs[0,0].plot(dados_vendas.loc['A'])
axs[0,0].set_title('Vendas da loja A')

axs[0,1].plot(dados_vendas.loc['B'])
axs[0,1].set_title('Vendas da loja B')

axs[1,0].plot(dados_vendas.loc['C'])
axs[1,0].set_title('Vendas da loja C')

axs[1,1].plot(dados_vendas.loc['D'])
axs[1,1].set_title('Vendas da loja D')

ymin = 0
ymax = 450

for ax in axs.ravel(): #Definindo o limite minimo e maximo dos graficos
  ax.set_ylim(ymin, ymax)

for ax in axs.flat: #Definindo uma grade para melhorar a visuaização em todos os graficos
  ax.grid()


plt.show()
```
